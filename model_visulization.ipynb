```python
import matplotlib.pyplot as plt

# Data from training process
epochs = [1, 2, 3, 4, 5]
train_loss = [0.0503, 0.0327, 0.0064, 0.0003, 0.0115]
val_loss = [0.0514, 0.0174, 0.0221, 0.0174, 0.0141]
accuracy = [0.9858, 0.9956, 0.9950, 0.9965, 0.9973]

# Plotting training and validation loss
plt.figure(figsize=(10, 5))
plt.plot(epochs, train_loss, label="Training Loss", marker='o')
plt.plot(epochs, val_loss, label="Validation Loss", marker='o')
plt.xticks(range(1, max(epochs) + 1))
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss Over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting accuracy
plt.figure(figsize=(10, 5))
plt.xticks(range(1, max(epochs) + 1))
plt.plot(epochs, accuracy, label="Validation Accuracy", marker='o', color='green')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Validation Accuracy Over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Training parameters
initial_lr = 3e-5
total_steps = 6905  # Total steps (from your training)
warmup_steps = int(0.1 * total_steps)

# Compute learning rates
lr_values = []
for step in range(total_steps):
    if step < warmup_steps:
        lr = initial_lr * (step / warmup_steps)
    else:
        lr = initial_lr * 0.5 * (1 + np.cos(np.pi * (step - warmup_steps) / (total_steps - warmup_steps)))
    lr_values.append(lr)

# Plot learning rate schedule
plt.plot(lr_values)
plt.title('Cosine Learning Rate Schedule with Warmup')
plt.xlabel('Steps')
plt.ylabel('Learning Rate')
plt.grid(True)
plt.show()
```
